# Exploratory Data Analysis

Q1. Make a Bar Graph for Gender

```python
a1 = sns.countplot( x = 'Gender', data=salesdata)

for bars in a1.containers:
    a1.bar_label(bars)
```

Q2. Which gender constitues to majority of Buyers and Which Gender has greater purchasing power

```python
gender_power = salesdata.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

a2 = sns.barplot( x= 'Gender', y="Amount", data=gender_power)
for bars in a2.containers:
    a2.bar_label(bars)
```

Q4. Find the Male and Female from different states

```python
a4 = salesdata.groupby(['State'], as_index=False)['Orders'].sum().sort_values( by = 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize': (20,5)})
sns.barplot(x ='State', y='Orders', data =a4)
```

Q5. total sales from top 10 states

```python
a5 = salesdata.groupby(['State'], as_index=False)['Amount'].sum().sort_values( by = 'Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize': (20,5)})
sns.barplot(x ='State', y='Amount', data =a5)
```

Q6. Find the No. of Sales as per Marital Status

```python
a6 = sns.countplot( data= salesdata, x = 'Marital_Status', hue = 'Gender')

for bars in a6.containers:
    a6.bar_label(bars)
```

Q6. Find the No. of Sales as per Occupation

```python
a6 = sns.countplot( data = salesdata, x= "Occupation")

sns.set(rc={'figure.figsize': (40,5)})
for bars in a6.containers:
    a6.bar_label(bars)
```

Q7. Find the No. of Sales for each product category

```python
a7 = sns.countplot(data = salesdata, x= 'Product_Category')

sns.set(rc={'figure.figsize': (50,10)})
for bars in a7.containers:
    a7.bar_label(bars)
```

Q8. Amount of Sales as per Product category

```python
a8 = salesdata.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values( by = 'Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize': (20,5)})
sns.barplot(x ='Product_Category', y='Amount', data =a8)
```

Q9. Sales as Per Product id and Sales Amount

```python
a8 = salesdata.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values( by = 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize': (20,5)})
a8a = sns.barplot(x ='Product_ID', y='Orders', data =a8)

for bars in a8a.containers:
    a8a.bar_label(bars)
```
